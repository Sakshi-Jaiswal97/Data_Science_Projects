```python
# Quartiles
# Calculate Quartiles

# Q1=25%
# Q2=50%
# Q3=75%
```

```python
Q1=data['Population in thousands (2017)'].quantile(0.25)
Q1

# NOTE: Percentile and Quartile  both does the same thing--> Thats why we get the same answer; Q1=q1
```

```python
Q2 = data['Population in thousands (2017)'].quantile(0.50)
Q2
```

```python
Q3= data['Population in thousands (2017)'].quantile(0.75)
Q3
```

```python
# IQR - Inter Quartile Range
# IQR = (Q3- Q1)

iqr = Q3-Q1
iqr
```

```python
# Calculate Upper fence
# Upper fence is above 75%.
# Upper fence : (Q3 + 1.5*IQR) --> Formula

uf = Q3 + (1.5*iqr)
uf
```

```python
# Calculate Lower fence
# Lower fence is below 0%.
# Lower fence :(Q1 - 1.5*IQR) --> Formula

lf= Q1 - (1.5*iqr)
lf
```

```python
# NOTE: Any value less than lower fence i.e -27712 and greater than upper fence i.e 47336 will be considered as an outliers.
# Outlier means values lies out side the range.
```

```python
# Code for outliers
# '|' --> this is "and" operator.

data[(data['Population in thousands (2017)']<lf) | (data['Population in thousands (2017)']>uf)]
```

```python
data[(data['Population in thousands (2017)']<lf) | (data['Population in thousands (2017)']>uf)].shape

# .shape --> Shows total number of rows and columns.
```

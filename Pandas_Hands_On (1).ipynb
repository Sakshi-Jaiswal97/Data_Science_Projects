**Q1.Given a sample list,tupe,and set with the elements as number from 1-10.Convert each of the datatype into pandas series.Create a seperate list of index values for the series.**

```python
list1 = list(range(1,11))
tuple1 = tuple(range(1,11))
set1 = set(range(1,11))

print(list1)
print(tuple1)
print(set1)

pd_list1 = pd.Series(list1)
pd_tuple1 = pd.Series(tuple1)
#pd_set1 = pd.Series(set1)  #set is unordered therefore it cannot directly converted into series because series assign index values.
#1st it has to convert into list and then into series.
# OR-->Typecast --> pd.Series([set1])

set1 = list(set1)
pd_set1 = pd.Series(set1)

print(pd_list1)
print(pd_tuple1)
print(pd_set1)

#Change index value from 1,2,3 (default index values) to a,b,c
index_value = ['a','b','c','d','e','f','g','h','i','j']
pd_list1 = pd.Series(list1, index = index_value)
print(pd_list1)

```

**Q11.Create two separate DataFrame A and B, and perform the following operations on them.**

**Q12.Create a DataFrame with column Name & Age.Perform the following operations:**

**(a).Write a sample code to search for the Name/Name where age is 20.**

**(b).Sort the dataframe in ascending order of the age.**

**(c).Using the python loop print the ages from the dataframe and store it in a set.**

```python
df = pd.DataFrame({'Name': ['a','b','c','d','e','f','g','h'],
                   'Age':[10,12,56,34,20,6,8,17]})
df
```

```python
#(a)

df[df['Age'] == 20]
```

```python
#(b)

df.sort_values(by = 'Age')  #ascending order (min to max)
```

```python
#(c)
list1 = []
for i in range(len(df)):
  list1.append(df.loc[i,'Age'])

print(list1)

set1 = set(list1)
print(set1)
```
